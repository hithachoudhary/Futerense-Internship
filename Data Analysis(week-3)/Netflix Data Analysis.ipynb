### Comparison of ratings from two countries

```python
#Ratings comparison of United States and India

US = df[df.country == "United States"]
order =  ['G', 'TV-Y', 'TV-G', 'PG', 'TV-Y7', 'TV-Y7-FV', 'TV-PG', 'PG-13', 'TV-14', 'R', 'NC-17', 'TV-MA']
plt.figure(figsize=(12,8))
g = sns.countplot(US.rating, hue=US.type, order=order, palette="dark:blue");
plt.title("Ratings for Movies & TV Shows in US")
plt.xlabel("Rating")
plt.ylabel("Total Count")
plt.show()
```

```python
IND = df[df.country == "India"]
order =  ['G', 'TV-Y', 'TV-G', 'PG', 'TV-Y7', 'TV-Y7-FV', 'TV-PG', 'PG-13', 'TV-14', 'R', 'NC-17', 'TV-MA']
plt.figure(figsize=(12,8))
g = sns.countplot(IND.rating, hue=IND.type, order=order, palette="dark:blue")
plt.title("Ratings for Movies & TV Shows in India")
plt.xlabel("Rating")
plt.ylabel("Total Count")
plt.show()
```
